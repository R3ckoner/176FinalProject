```python
import pandas as pd
import matplotlib.pyplot as plt
from mplfinance.original_flavor import candlestick_ohlc
import matplotlib.dates as mdates
import seaborn as sns
import io
from google.colab import files

# Upload the file
uploaded = files.upload()

# Load the dataset
for filename in uploaded.keys():
    data = pd.read_csv(io.BytesIO(uploaded[filename]))

# Convert 'End' dates to datetime format
data['End'] = pd.to_datetime(data['End'])

# Sort data by 'End' dates
data.sort_values(by='End', inplace=True)

# Line Chart of Bitcoin Prices Over Time
plt.figure(figsize=(12, 6))
plt.plot(data['End'], data['Close'], color='blue')
plt.title('Bitcoin Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.show()

# Candlestick Chart of Daily Bitcoin Prices
plt.figure(figsize=(12, 6))
ax = plt.subplot()
candlestick_ohlc(ax, data[['End', 'Open', 'High', 'Low', 'Close']].values, width=0.6, colorup='green', colordown='red')
ax.set_title('Daily Bitcoin Prices')
ax.set_xlabel('Date')
ax.set_ylabel('Price (USD)')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Scatter Plot Matrix
sns.pairplot(data[['Open', 'High', 'Low', 'Close', 'Volume', 'Market Cap']])
plt.suptitle('Scatter Plot Matrix of Bitcoin Features', y=1.02)
plt.show()
```
